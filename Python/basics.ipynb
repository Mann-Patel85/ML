```python
def max(*args,**kwargs):
    print("*Hello")
    for i in args:
        print(i)
    
    print("**Hello")
    for x,y in kwargs.items():
        print(f"{x}=={y}")
        
max('Welcome','Home',Name='Mann',Age=21)
```
